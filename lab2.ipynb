```python
#creating a df
import pandas as pd
d1={'Name':['jai','pavan','simon','manohar'],
    'Age':[20,24,24,25],
    'Address':['banglore','chennai','mysore','kerala'],
    'Qualification':['Mca','Mba','mtech','Mca']}

d2={'Name':['pavansj','praveen','mohan','dj'],
    'Age':[22,28,23,29],
    'Address':['andhra','mumbai','delhi','Nagpur'],
    'Qualification':['bca','phd','Bcom','mba']}

df = pd.DataFrame(d1,index=[0,1,2,3])
df1 = pd.DataFrame(d2,index=[4,5,6,7])


print(df,"\n\n\n\n", df1)
```

```python
#concat
frames=[df,df1]
re1=pd.concat(frames)
re1
```

```python
#concating fd by ignoring indexes
import pandas as pd
d1={'Name':['jai','pavan','simon','manohar'],
    'Age':[20,24,24,25],
    'Address':['banglore','chennai','mysore','kerala'],
    'Qualification':['Mca','Mba','mtech','Mca'],
   'mobile no':[95,88,66,44]}

d2={'Name':['pavansj','praveen','mohan','dj'],
    'Age':[22,28,23,29],
    'Address':['andhra','mumbai','delhi','Nagpur'],
    'Qualification':['bca','phd','Bcom','mba'],
   'Salary':[1000,2000,3000,9999]}

df = pd.DataFrame(d1,index=[0,1,2,3])
df1 = pd.DataFrame(d2,index=[3,5,6,7])


print(df,"\n\n\n\n", df1)
```

```python
res=pd.concat([df,df1], ignore_index=True)
res
```

```python
#concating df with group keys
frames=[df,df1]
res=pd.concat(frames,keys=['x','y'])
res
```

```python
#concatenating with ndims
import pandas as pd
d1={'Name':['jai','pavan','simon','manohar'],
    'Age':[20,24,24,25],
    'Address':['banglore','chennai','mysore','kerala'],
    'Qualification':['Mca','Mba','mtech','Mca']}
df=pd.DataFrame(d1,index=[0,1,2,3])
s1=pd.Series([1000,2000,3000,9999],name='Salary')
print(df,"\n\n",s1)
```

```python
res=pd.concat([df,s1],axis=1)
res
```

```python
#merging df with one unique key combination
d1={'key':['k0','k1','k2','k3'],
   'Name':['jai','pavan','simon','manohar'],
    'Age':[20,24,24,25]}
d2={'key':['k0','k1','k2','k3'],
    'Name':['pavansj','praveen','mohan','dj'],
    'Age':[22,28,23,29]}
df=pd.DataFrame(d1)
df1=pd.DataFrame(d2)
print(df,"\n,\n",df1)
```

```python
#merge
res=pd.merge(df,df1,on='key')
res
```

```python
#merging df using multiple join keys
d1={'key':['k0','k1','k2','k3'],
    'key1':['k0','k1','k0','k1'],
   'Name':['jai','pavan','simon','manohar'],
    'Age':[20,24,24,25]}
d2={'key':['k0','k1','k2','k3'],
    'key1':['k0','k0','k0','k0'],
    'Address':['banglore','chennai','mysore','kerala'],
    'Qualification':['Mca','Mba','mtech','Mca']}
df=pd.DataFrame(d1)
df1=pd.DataFrame(d2)
print(df,"\n,\n",df1)
```

```python
res1=pd.merge(df,df1,on=['key','key1'])
res1
```

```python
#merging df using how in an argument
d1={'key':['k0','k1','k2','k3'],
    'key1':['k0','k1','k0','k1'],
   'Name':['jai','pavan','simon','manohar'],
    'Age':[20,24,24,25]}
d2={'key':['k0','k1','k2','k3'],
    'key1':['k0','k0','k0','k0'],
    'Address':['banglore','chennai','mysore','kerala'],
    'Qualification':['Mca','Mba','mtech','Mca']}
df=pd.DataFrame(d1)
df1=pd.DataFrame(d2)
print(df,"\n,\n",df1)
```

```python
#left merge 
res=pd.merge(df,df1,how='left',on=['key','key1'])
res
```

```python
#right merge
res=pd.merge(df,df1,how='right',on=['key','key1'])
res
```

```python
#outer merge
res2=pd.merge(df,df1,how='outer',on=['key','key1'])
res2
```

```python
#inner merge
re3=pd.merge(df,df1,how='inner',on=['key','key1'])
re3
```

```python
#joining df
d1={
    'Name':['jai','pavan','simon','manohar'],
    'Age':[20,24,24,25]
   }
d2={
    'Address':['andhra','mumbai','delhi','Nagpur'],
    'Qualification':['bca','phd','Bcom','mba']
}
df=pd.DataFrame(d1, index=['k0','k1','k2','k3'])
df1=pd.DataFrame(d2, index=['k0','k2','k3','k4'])
print(df,"\n\n",df1)
```

```python
res=df.join(df1)
res
```

```python
res1=df.join(df1,how='outer')
res1
```
