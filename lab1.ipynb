```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({"A":[12.0,4,5,None,1],
                "B":[None,2,54,3,None],
                "C":[20,16,51,3,8],
                "D":[14,3,None,None,8]})
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
df1=df.fillna(0)
df1
```

```python
df2=df.fillna(method='pad')
df2
```

```python
# filling null values with the next values
df3=df.fillna(method='bfill')
df3
```

```python
# to compute mean for the values
fillmean=df3.fillna(df3.mean())
fillmean
```
